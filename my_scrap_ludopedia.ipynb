# Captando a página inteira

```python
cards_livros = []
```

```python
todos_livros = soup.findAll('div' , {'class' : "pad-top"})
```

```python
todos_livros
```

```python
for livro in todos_livros:
  cards_novos = {}
  titulo = livro.find('h4', {'class': 'media-heading'}).getText()
  notas = livro.find('div', {'class': 'rank-info'})
  lista_notas = notas.findAll('span')
  nota_rank = lista_notas[0].getText()
  media = lista_notas[1].getText()
  quantidade_notas = lista_notas[2].getText()
  cards_novos['titulo'] = titulo
  cards_novos['nota_rank'] = nota_rank.split()[2]
  cards_novos['media'] = media.split()[1]
  cards_novos['quantidade_notas'] = quantidade_notas.split()[1]
  cards_livros.append(cards_novos)
```

```python
livros_ludopedia = pd.DataFrame(cards_livros)
```
